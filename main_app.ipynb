```python
import pandas as pd
import joblib

# Load the cleaned data
data = pd.read_csv("cleaned_fitness_data.csv")

# Load your trained model
model = joblib.load("improved_fitness_predictor.pkl")

# Drop columns the model wasn’t trained on
for col in ['date', 'fitness_level']:
    if col in data.columns:
        data = data.drop(col, axis=1)

# Make predictions
predictions = model.predict(data)

# Add predictions to the dataframe
data['Predicted_Fitness_Level'] = predictions

# Save output
data.to_csv("predicted_fitness_output.csv", index=False)

print("✅ Predictions added! Check 'predicted_fitness_output.csv' for results.")
```
